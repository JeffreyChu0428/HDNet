```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']  # 設定字型為微軟正黑體
plt.rcParams['axes.unicode_minus'] = False  # 正確顯示負號

# 模型條件
categories = ['本模型', '無HDC分類器', '無二階段架構', '無個人化']

# 各項指標資料
overall_accuracy = [0.822, 0.732, 0.735, 0.739]
f1_scores = [0.730, 0.620, 0.656, 0.674]
normal_accuracy = [0.783, 0.700, 0.624, 0.571]

# x 軸位置與條寬
x = range(len(categories))
bar_width = 0.25

# 建立圖表
plt.figure(figsize=(12, 5))

# 畫三組條
bars1 = plt.bar([i - bar_width for i in x], overall_accuracy, width=bar_width, label='整體準確率')
bars2 = plt.bar(x, f1_scores, width=bar_width, label='F1-score')
bars3 = plt.bar([i + bar_width for i in x], normal_accuracy, width=bar_width, label='正常類別準確率')

for bars in [bars1, bars2, bars3]:
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, f"{yval:.3f}", ha='center', va='bottom', fontsize=9)

# 圖表設定
plt.xticks(x, categories)
plt.ylabel("指標數值")
plt.title("各模組對分類效能指標之影響")
plt.ylim(0, 1)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# 顯示圖表
plt.show()
```
